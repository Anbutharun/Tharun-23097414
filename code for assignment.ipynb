```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Supermart Grocery Sales - Retail Analytics Dataset.csv')
```

```python
print("Dataset Overview:\n", df.info())
```

```python
print("\nFirst 5 Rows of the Dataset:\n", df.head())
```

```python
null_values = df.isnull().sum()
print("Null values in each column:\n", null_values)
```

```python
data_types = df.dtypes
print("Data Types in the Dataset:\n", data_types)
```

Visulization

Bar Plot of Sales by Category

```python
def plot_total_profit_by_category(df):
    # Grouping the data by Category and summing the Profit
    category_profit = df.groupby('Category')['Sales'].sum().reset_index()
    # Creating a bar plot
    plt.figure(figsize=(8, 6))
    sns.barplot(x='Category', y='Sales', data=category_profit, palette='viridis')
    plt.title('Total Sales by Category')
    plt.xlabel('Category')
    plt.ylabel('Sales')
    plt.xticks(rotation=45)
    plt.show()
plot_total_profit_by_category(df)
```

Scatter Plot of Sales vs. Profit

```python
def plot_sales_vs_profit_scatter(df):
    plt.figure(figsize=(10, 6))
    plt.scatter(df['Sales'], df['Profit'], alpha=0.5, color='orange')
    plt.title('Sales vs. Profit')
    plt.xlabel('Sales Amount')
    plt.ylabel('Profit Amount')
    plt.grid()
    plt.show()
plot_sales_vs_profit_scatter(df)
```

Box plot of Profit by Category

```python
def plot_profit_boxplot(df):
    plt.figure(figsize=(12, 6))
    sns.boxplot(x='Category', y='Profit', data=df)
    plt.title('Box Plot of Profit by Category')
    plt.xlabel('Category')
    plt.ylabel('Profit Amount')
    plt.xticks(rotation=45)
    plt.show()
plot_profit_boxplot(df)
```
